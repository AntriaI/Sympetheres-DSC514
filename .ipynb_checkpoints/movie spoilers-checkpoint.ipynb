```python
import pandas as pd
df = pd.read_csv ("movie_spoiler_sample.csv")
df
```

```python
df = df.iloc[:, :13]  # Keep only the first 13 columns
```

```python
#Genre column - identify and remove erroneous values, split and perform encoding
print(df["genre"].value_counts())
```

```python
# Drop rows where 'genre' does NOT start with '['  - ie any invalid values
df = df[df['genre'].str.startswith("[", na=False)]

# split the genres and remove the "[" "]" characters
df["genre"]=df["genre"].str.split(",")

df_exploded = df.explode('genre')
df_exploded['genre'] = df_exploded['genre'].str.replace("[\[\]]", "", regex=True)

# Apply one-hot encoding using get_dummies
df_encoded = pd.get_dummies(df_exploded['genre'])

# merge with the original dataframe
df_final = df.merge(df_encoded.groupby(df_exploded.index).sum(), left_index= True, right_index = True)

#  Drop the original 'genre' column if it's no longer needed
df_final = df_final.drop(columns=['genre'])
df =df_final
df
```

```python
#check for NAs in target variable
df[df["is_spoiler"].isna()]
#drop these rows since they do not help train the model
df = df.dropna(subset=['is_spoiler'])

#now check if there are any other NAs
df[df.isna().any(axis=1)]
# there are no more missing values left
```

```python
#check if the  target variable is balanced
df["is_spoiler"].value_counts()
#target variable is balanced 
```
