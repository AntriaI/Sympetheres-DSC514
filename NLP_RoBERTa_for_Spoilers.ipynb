```python
from transformers import AutoTokenizer
```

```python
# RoBERTa tokenizer
tokenizer = AutoTokenizer.from_pretrained("roberta-large")
```

```python
# Tokenizing the text data
df_roberta['tokenized_text'] = df_roberta['review_text'].apply(lambda x: tokenizer(x, padding='max_length', truncation=True, max_length=512, return_tensors="pt"))

# Parameters: The tokenizer is a pre-trained tokenizer for RoBERTa that takes in a string (x) and converts it into numerical tokens that the model can understand.
# It ensures that all tokenized sequences are padded to the same length (here 512 max length).
# Padding adds special tokens to make shorter sequences the same length as the longest allowed sequence.
# "pt" -> Converts the output into PyTorch tensors
```

We want to transform the dataset in order to fine-tune the Roberta model:

```python
from torch.utils.data import Dataset
```

```python
# This is a custome dataset class

class SpoilerDataset(Dataset):
    def __init__(self, texts, labels):
        self.texts = texts
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            'input_ids': self.texts[idx]['input_ids'].squeeze(0),
            'attention_mask': self.texts[idx]['attention_mask'].squeeze(0),
            'labels': torch.tensor(self.labels[idx], dtype=torch.long)
        }

# Dataset preparation

texts = df_roberta['tokenized_text']
labels = df_roberta['is_spoiler_binary'].tolist()

dataset = SpoilerDataset(texts.tolist(), labels)
```

Fine-Tuning Roberta:

```python
from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments
```

```python
# RoBERTa model for binary classification
model = AutoModelForSequenceClassification.from_pretrained("roberta-large", num_labels=2)
```

```python
# Training arguments
training_args = TrainingArguments(output_dir="./results",evaluation_strategy="epoch",learning_rate=2e-5,per_device_train_batch_size=8,per_device_eval_batch_size=16,num_train_epochs=3,weight_decay=0.01,logging_dir="./logs",)
```

```python
# Splitting the dataset
from torch.utils.data import random_split
train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])
```

```python
# Training
trainer = Trainer(model=model,args=training_args,train_dataset=train_dataset,eval_dataset=test_dataset,tokenizer=tokenizer,)
trainer.train()
```

```python
# ?????????????????
def predict(text):
    tokens = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    tokens = {key: val.to("cuda") for key, val in tokens.items()}  # Move tokens to GPU
    outputs = model(**tokens)
    prediction = torch.argmax(outputs.logits, dim=1).item()
    return "Spoiler" if prediction == 1 else "Not a Spoiler"

print(predict("This movie reveals the shocking death of the hero!"))
```
